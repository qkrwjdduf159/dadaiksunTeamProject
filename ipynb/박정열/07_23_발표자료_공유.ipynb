## 1. 데이터 증강

### 데이터 증강을 생각한 이유

- 1. 데이터의 양이 너무 적다.
- 2. 예측을 한 결과 각 머신러닝에서 rmse, score, smape이 좋지 못한 결과가 나왔다.

#### 데이터 증강에 사용한 방법|

##### 보간법(Interpolate)
- @김선홍(진행중)

##### GAN (Generative Adversarial Network)
- @박정열(삐그덕)

##### SMOTE Regression
- @김선홍@박정열
- SMOTE Regression을 해본 결과 데이터의 증강을 할 때 imbalance의 데이터를 증가시켜주는 경향이 있었고 그 부분은 저희 데이터에 적용시켜 봤을 때 평가 척도인 score, rmse, smape에 부정적 영향을 미쳤기 때문에 쓰지 않습니다.

###### tsaug(Time Series data augmentation)
- @박정열
- tsaug는 target변수와 time series만 넣어주고 증강을 시켜주는 방식이여서 모든 데이터를 증강해야하는 저희 데이터에는 맞지 않아서 쓰지 않습니다.

# 2. 다른 모델의 도입

## LMM(Linear Mixed Model)

- 정의: 고정효과(fixed effect)와 랜덤(임의)효과(random effect)로 이루어진 모형

- 고정효과 : 모든 샘플과 모집단에서 똑같이 고정되어 있는 요인의 효과
- 랜덤효과 : 모집단으로부터 fixed되지 않은 요인의 효과

#### Linear Mixed Model calculation
##### y = Xb + Zu + e 

y는 형질을 관측한 값으로 이루어진 벡터이다.

b는 고정효과로 알려진 요인들의 벡터이다.

u는 랜덤효과라고 알려진 요인들의 벡터이다.

e는 잔차항의 벡터로 역시 임의효과이다.
